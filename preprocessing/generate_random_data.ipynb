```python
import pandas as pd
import numpy as np
```

### load dataset to be modeled after

```python
df = pd.read_pickle("../data/data_100hz_intrasubject_scaled_expanded.pkl")

labels = ["episode_type","trial_type","trial_num",
          "episode_num","is_corrupt","action_index",
          "trimmed_data","subject_num","left","win"]

eeg_df = df.drop(labels, axis=1)
labels_df = df[labels]
labels_df
```

```python
eeg_df.describe()
```

### create randomized eeg data dataframe of same shape as original data

```python
random_eeg_data = (np.random.rand(eeg_df.shape[0], eeg_df.shape[1])-.5)*30
random_eeg_data_df = pd.DataFrame(random_eeg_data)

random_eeg_data_df
```

```python
df_ = random_eeg_data_df.join(labels_df)

df_.describe()
```

```python
df_.to_pickle("../data/RANDOM_data_100hz_intrasubject_scaled_expanded.pkl")
```
