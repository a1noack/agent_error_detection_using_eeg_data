```python
import pandas as pd
```

### load the data and separate into eeg data and labels on a per episode basis

```python
df = pd.read_pickle("data/data_final.pkl")

labels = ["episode_type","trial_type","trial_num",
          "episode_num","is_corrupt","action_index",
          "trimmed_data","subject_num","left","win"]

eegs = df.drop(labels, axis=1)
labels_df = df[labels]

eegs.head()
```

### only keep a certain portion of each episode's readings

```python
#only using the eeg readings taken after Pacman's decision
eegs = eegs[list(range(1024,1860))]

eegs.head()
```

### rejoin the cropped eeg data and the original labels

```python
df_new = eegs.join(labels_df)

df_new.head()
```

### save to pickle

```python
df_new.to_pickle("data/data_halved.pkl")
```
