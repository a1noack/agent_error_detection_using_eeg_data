```python
import pandas as pd
```

```python
df = pd.read_pickle("../data/data_100hz_intrasubject_scaled.pkl")

labels = ["episode_type","trial_type","trial_num",
          "episode_num","is_corrupt","action_index",
          "trimmed_data","subject_num","left","win"]
```

### specify the cropping/expanding strategy

```python
#how far into the readings to slice
depth = 60
#how often to create a new copy
crop_stride = 2 #once every 5 readings

#number of eeg readings per episode in orginal dataset
orig_len = 375
#calculate new length of eeg readings
new_len = orig_len - depth

#placeholder dataframe for expanded data
df_ = pd.DataFrame()
```

### loop and update the placeholder dataframe

```python
for i in range(0, depth, crop_stride):
    #create a list of the readings to keep in this iteration
    eeg_keep = list(range(i, new_len+i))
    
    #add in the labels to that list
    columns_keep = eeg_keep+labels
    
    #keep the resulting dataframe
    df_i = df[columns_keep]
    
    #rename the columns in the resulting dataframe so that they all line up, index-wise
    df_i.columns = list(range(new_len))+labels
    
    #append the iteration dataframe to the placeholder
    df_ = df_.append(df_i)
```

### display and save new_df

```python
df_ = df_.reset_index(drop=True) #9288 = (40/5)*1161
```
